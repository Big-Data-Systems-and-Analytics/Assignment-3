```python
import zipfile
from PIL import Image
import matplotlib.pyplot as plt

def image_gen():
    imgzip = zipfile.ZipFile("archive.zip")
    inflist = imgzip.infolist()

    for f in inflist:
        ifile = imgzip.open(f)
        img = Image.open(ifile)
        #img.show()
        #print(img)
        plt.figure(figsize= (12,12))
        plt.imshow(img)

image_gen()
```
